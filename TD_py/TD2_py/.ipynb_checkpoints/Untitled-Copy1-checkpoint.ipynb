```python
import numpy as np;
import math;
import random;
import matplotlib.pyplot as plt;
import pandas as pd;
```

```python
def tirage_pf(n,p):
    # Effectue n tirage a Pile (P) ou face (F) (1/2,1/2)
    X="";
    for i in range(n):
        U=random.random();
        if U <= p:
            X=X+'P'
        else:
            X=X+'F'
    return X

def max_length(U):
# Calcule le nombre maximum de P consecutifs
# dans la suite U
    MAX=0;N=0;
    for n in range(len(U)):
        if U[n]=='P':
            N=N+1
        else:
            N=0
        MAX=max(MAX,N);
    return MAX
```

```python
def histo_discret(samples, maxsize):
# histogramme de tirages selon une loi discrète à valeurs entières
# supposé prendre des valeurs entre 0 et max.
  H=np.zeros(maxsize+1)
  for k in range(maxsize+1):
    # Calcul du nbre de tirages valant k / Taille
    H[k]=float(np.size(np.nonzero(samples==k))) / np.size(samples); 
  return H;
```

```python
def main_1():
    p=1.0/2.0;
    # On teste cette fonction avec N=20
    # rand(1:20) renvoie 100 tirages uniformes sur l'intervalle [0,1]
    U=tirage_pf(20,p)# 20 tirages a pile ou face 1/2,1/2  (1=Pile,0=Face)
    max_length(U)# nombre mximum de P consecutifs

    # On effectue 1000 tirages avec N=100
    N=100;
    Taille=100000;# nbre de simulation
    X=np.zeros(Taille);
    for i in range(Taille):
        U=tirage_pf(N,p);
        X[i]=max_length(U);
    size, scale = 10000, 10
    my_histo = pd.Series(X)
    my_histo.plot.hist(grid=True, bins=20, rwidth=0.9, color='#607c8e')
    plt.title('Loi de la longueur maximale pour 10 000 simulation de longueur 100')
    plt.xlabel('Loi de la longueur maximale')
    plt.ylabel(' ')
    plt.grid(axis='y', alpha=1)
```

```python
main_1()
```

```python
def trajectoire(U):
# On calcule la trajectoire de X
    X=np.zeros(len(U)+1);
    val=0;
    for n in range(len(U)):
        if U[n]=='P':
            val=val+1;
        else:
            val=0;
        X[n+1]=val;
    return X;
```

```python
def main_2():
  N=100;
  p=1.0/2.0;

  # On trace une trajectoire de X
  U=tirage_pf(N,p);
  X=trajectoire(U);
  plt.plot(X);

  # 4 trajectoires
  for i in range(0):
    U=tirage_pf(N,p);
    X=trajectoire(U);
    plt.plot(X);
    
main_2()
```

```python
def proba(N,l,p):
    # Calcule la probabilite de voir au moins l piles consecutifs
    # dans N tirages a pile (p) ou face (1-p)

    # la matrice de transition de la chaîne 
    # de taille (l+1,l+1)
    P=np.zeros((l+1,l+1))
    P[0:l,0]=1-p
    P[0:l,1:l+1]=p*np.eye(l)
    P[l,l]=1
    PN=np.linalg.matrix_power(P,N)
    return PN[0,l]
```

```python
def calculer_loi(N,p):
  MAXMAX=50;# a choisir assez grand (mais pas trop ...)
  loi=np.zeros(MAXMAX+1)
  
  previous=1;# proba d'avoir au moins 0 pile = 1 !
  # le support de la loi est [0,1,...,N] que l'on tronque en MAXMAX
  for l in range(min(N,MAXMAX)+1):
    current=proba(N,l+1,p); # proba d'avoir au moins l+1 pile
    loi[l]=previous - current;# proba d'avoir exactement l pile
    previous=current;
  return loi
```

```python
def main_3():
    # On teste avec N=1 et N=2, p=1/2
    # Pour N=1, 0 pile avec proba 1/2 et 1 pile avec proba 1/2
    calculer_loi(1,1/2.0)
    # Pour N=2, on doit trouver (1/4,1/2,1/4) pour (0,1,2)
    calculer_loi(2,1/2.0)
    # en principe ca marche ...
  
    N=100;p=1.0/2.0;
    loi=calculer_loi(N,p);
    sum(loi) # on verifie que ca somme a 1
  
    # dessin
    plt.bar(range(20), loi[0:20])
   
    # comparaison avec les simulations
    Taille=10000;
    X=np.zeros(Taille);
    for i in range(Taille):
        U=tirage_pf(N,p);
        X[i]=max_length(U);
  
    histo=np.zeros(21);
    for i in range(21):
        histo[i]=np.size(np.where(X==i))/(1.0*Taille);
    epsilon=np.linalg.norm(loi[0:20]-histo[0:20])
    # doit etre "petit", pour bien faire il faudrait faire un
    # test du |$\xi^2$| pour savoir ce que "petit" veut dire ici.
    print("epsilon = ",epsilon," --- petit en principe.")
```

```python
main_3()
```

```python
def simul_cox_ross(N,x_0,p,u,d):
  U=np.random.binomial(1,p,N);# tirages a pile ou face (p,1-p)
  X=np.zeros(np.size(U));
  X[0]=x_0;
  for i in range(np.size(U)-1):
    if U[i+1]==0: 
        X[i+1]=X[i]*u
    else:
        X[i+1]=X[i]*d;
  return X
```

```python
K=100;
def f(x):
  return max(x-K,0);
#-------------------------------------------------------
def prix_recursif(x,k,N):
    if (k==N):
        return f(x)
    else:
        return p*prix_recursif(x*u,k+1,N)+(1-p)*prix_recursif(x*d,k+1,N)
#-------------------------------------------------------
def prix_slow(x,N):
  return prix_recursif(x,0,N);
```

```python
def inc(n):
# Permet de faire comme si les tableaux était indicés 
# à partir de 0 (et non de 1)
 return n+1;
#-------------------------------------------------------
def prix(x_0,N):
    # U=zeros(inc(N),inc(N))
    U=np.zeros([N+1,N+1])
    for k in range(N+1):
        U[N,k] = f(x_0 * pow(u,k) * pow(d,N-k))
    for n in range(N-1,-1,-1): # [N-1:-1:0] le temps decroit de N-1 a 0
        for k in range(0,n+1): # [0:n]
            U[n,k] = p*U[n+1,k+1]+(1-p)*U[n+1,k]
    return U[0,0]
```

```python
def liste_moyenne_rec(x,s,k,N):
# On constitue la liste des points visités par la chaine
# Si un point est visité deux fois, il y figure 2 fois.
    # print k, N, x, s
    if (k==N):
        liste=[];
        liste.append([x,s]);
        return liste
    else:
        liste_up = liste_moyenne_rec(x*u,s+x*u,k+1,N);
        liste_down = liste_moyenne_rec(x*d,s+x*d,k+1,N);
        liste_up.extend(liste_down);
        return liste_up;
    
def liste_moyenne(x,N):
  # On part de (x,s=x) a l'instant 0
  return liste_moyenne_rec(x,x,0,N)

def main_10():
  x_0=100
  N=10
  liste=liste_moyenne(x_0,N)

  # Tri des points selon les valeurs de la somme.
  # Les valeurs de x peuvent etre egales, mais pas celle de s.
  # Nous allons le verifier.
  liste.sort();

  # On regarde si tous les points sont differents
  # en parcourrant le tableau ainsi classé
  epsilon=0.001
  Taille=len(liste)
  match=[]
  for i in range(Taille-1):
    if (np.linalg.norm(np.asarray(liste[i]) - np.asarray(liste[i+1])) < epsilon):
      print ('Warning: (', liste[i][0],',',liste[i][1], ') ~ (',liste[i+1][0],',',liste[i+1][0],')\n')
      match.append([liste[i][1],liste[i][2],liste[i+1][1],liste[i+1][2]]);
                                      
  if len(match) == 0: 
     print ("Aucun point n'est dupliqué.\n")
        
main_10()
```
